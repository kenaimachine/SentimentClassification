````python
import pathlib
import re
path = data_dir_path+"dataset/train/pos/"

def check_match(mObj):
    if mObj is not None:
        matched = mObj.group(0) 
        return matched
    else:
        return 'Not_Matched'


def getText(path):
    txtlist=[]
    all_files=os.listdir(path)
    r=re.compile(r"\d{,2}_[0-9]+\.txt")
    
    pathMatch=[check_match(r.match(i)) for i in all_files]
    pathMatch=list(filter(lambda x: x!='Not_Matched', pathMatch))
    
    def getDigit(text):
        return int(text) if text.isdigit() else text
    
    def getKey(text):
        return [getDigit(c) for c in re.split(r'_',text)]
    
    print(pathMatch.sort(key=getKey),len(pathMatch)) 
    print(pathMatch)
    
    for p in pathMatch:
        #print(p)
        with open(path+p,'r') as current_file:
                txtlist.append(current_file.read())
                current_file.close()
    return txtlist     


train_pos=pd.Series(getText(path), dtype='object')

print(train_pos)
pos = pd.DataFrame({"reviews": train_pos, "labels": list(np.ones(len(train_pos), dtype=int))})
pos.head()
````

##Task 8 Remove Stop words
Use the stop word in nltk   
stop_words = set(stopwords.words('english'))    
Based the set of stop_words remove the stop words in the train and test dataset Reviews column

````python
def remove_stop_words(dataset):
    stop_words = set(stopwords.words('english'))
  #add code
    dataset.reviews=dataset.reviews.apply(lambda word: [x.lower() for x in word if x not in stop_words])  
    dataset.reviews=dataset.reviews.apply(lambda wordlist: [x.strip() for x in wordlist])
    return dataset
````

````python
train_data = remove_stop_words(train_data)
test_data = remove_stop_words(test_data)
print(train_data.head())
````

##Task 9 Lemmatization  
https://www.geeksforgeeks.org/python-lemmatization-with-nltk/    

Resolving words to their dictionary form    
lemmatizer = nltk.stem.WordNetLemmatizer()    

From the tokenization and remove stop each reviews data in the dataframe is a list of words.     
In the lemmatization process join the list of words into a string and store into the reviews column in the dataframe


````python
from nltk.stem import WordNetLemmatizer


def normalize(dataset):
    
    lemmatizer = WordNetLemmatizer() 
    #w=dataset.reviews=dataset.reviews.apply(lambda word: [lemmatizer.lemmatize(i)==i for i in word]) 
    dataset.reviews=dataset.reviews.apply(lambda word: ' '.join(str(i) for i in [lemmatizer.lemmatize(i)for i in word])) 
    #print(w)

    return dataset
````

````python
train_data = normalize(train_data)
test_data = normalize(test_data)
print(train_data.head(10))
````

##Task 10 Remove unwanted symbols
Given  garbage = "~`!@#$%^&*()_-+={[}]|\:;'<,>.?/"    
Remove the unwanted symbols in the train and test dataframe reviews column


````python
def remove_garbage(dataset):
    garbage = "~`!@#$%^&*()_-+={[}]|\:;'<,>.?/"
     
    r=re.compile(r"[^~`!@#$%\^&*()_\-+={[}\]|\\:;'<,>.?/]*")    
    
    #r=re.compile(r"[~`!@#$%\^&*()_\-+={[}\]|\\:;'<,>.?/]*")     # to test if removal works.


    #dataset.reviews=dataset.reviews.apply(lambda elem:[r.match(i).group() for i in elem])
    dataset.reviews=dataset.reviews.apply(lambda elem:' '.join(r.findall(elem)).strip())        
        
    return dataset
````

````python
from sklearn.feature_extraction.text import TfidfVectorizer
tr=train_data.copy()
ts=test_data.copy()

def fit_corpus(train_data, test_data):
  #add code
  #Pass in the train_data and test_data dataframe    
  #Extract both dataframe Reviews column    
  #Combine into one dataframe   
  #Define tfidf = TfidfVectorizer()    
  #Fit the dataframe tfidf.fit(df["reviews"])
  #

    df= pd.concat([train_data, test_data],ignore_index=True )
    #display(df.head(), df.tail())
    tfidf = TfidfVectorizer(ngram_range=(1,2),min_df=2,max_df=0.5)  

    
    #print(df.reviews.apply(lambda x: ','.join(x)))
    #df.reviews=df.reviews.apply(lambda x: ','.join(x))
    #display(df['reviews'])
    
    tfidf.fit(df['reviews'])
    #print(len(tfidf.get_feature_names()))
    #print(tfidf.get_feature_names())
    
    return tfidf

tfidf = fit_corpus(tr,ts)
#print(len(tfidf.vocabulary_),tfidf.vocabulary_) 
#print(tfidf.get_feature_names())

def transform_data(tfidf, dataset):
    #print(dataset.reviews.head())
    #dataset.reviews=dataset.reviews.apply(lambda x:','.join(x))
    
    features = tfidf.transform(dataset["reviews"])
    #print(features)
    return pd.DataFrame(features.todense(), columns = tfidf.get_feature_names())

#print(len(tfidf.vocabulary_),tfidf.vocabulary_) 
train_features = transform_data(tfidf, tr)

pd.set_option('display.max.columns',None)


train_features.head()
display(train_features.shape, train_features.head())

````

````python
tr=train_data.copy()
ts=test_data.copy()
test_features = transform_data(tfidf, ts)
````

````python
train_labels = tr["labels"]
test_labels = ts["labels"]
print(train_labels.head(),test_labels.head())
````
